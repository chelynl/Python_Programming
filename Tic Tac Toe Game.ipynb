## Assignment 2: Tic-Tac-Toe

### Objectives:
1. Implement a program that tracks the state of an ongoing tic-tac-toe game.
2. Allow each player to choose the position they want to place their pieces.
3. Check and end the game when either a win or a tie occurs.

### Step-by-step Guide:
- Initialize the board to be empty.
- X starts, and chooses a position to place their marker.
- O positions their marker in any empty position.
- After each marker is placed, a check is made to see whether either player has won.
- Placement alternates until a player wins or the board is full, at which point the game is called for a tie.

```python
"""
Create a dictionary called board_temp with keys marking positions on a tic-tac-toe board.
The board is initially empty and it will get filled with Xs and Os based on positions chosen by each player.

"""

board_temp = {
    'A1': ' ' , 'B1': ' ' , 'C1': ' ' ,
    'A2': ' ' , 'B2': ' ' , 'C2': ' ' ,
    'A3': ' ' , 'B3': ' ' , 'C3': ' ' 
}

# Store position keys into a list
board_keys = []

for key in board_temp:
    board_keys.append(key)

# Create a function that shows the current board to keep track of the game status
def show_board(board):
    print(board['A1'] + '|' + board['B1'] + '|' + board['C1'])
    print('-+-+-')
    print(board['A2'] + '|' + board['B2'] + '|' + board['C2'])
    print('-+-+-')
    print(board['A3'] + '|' + board['B3'] + '|' + board['C3'])

# Now we'll write the main function which has all the gameplay functionality.
def play_game():

    turn = 'X'
    count = 0


    for i in range(10):
        show_board(board_temp)
        print("It's your turn, " + turn + ". Choose your position:")

        position = input()
        position = position.upper()

        if position not in board_keys:
            print("That is not a valid position. Pick again.")
            continue
        elif board_temp[position] == ' ':
            board_temp[position] = turn
            count += 1
        else:
            print("That place is already taken.")
            continue

        # Now we will check if player X or O has won,for every move after 5 moves. 
        if count >= 5:
            if board_temp['A1'] == board_temp['B1'] == board_temp['C1'] != ' ': # across the top
                show_board(board_temp)
                print("\nGame Over.\n")                
                print(turn + " won.")                
                break
            elif board_temp['A2'] == board_temp['B2'] == board_temp['C2'] != ' ': # across the middle
                show_board(board_temp)
                print("\nGame Over.\n")                
                print(turn + " won.")
                break
            elif board_temp['A3'] == board_temp['B3'] == board_temp['C3'] != ' ': # across the bottom
                show_board(board_temp)
                print("\nGame Over.\n")                
                print(turn + " won.")
                break
            elif board_temp['A1'] == board_temp['A2'] == board_temp['A3'] != ' ': # down the left side
                show_board(board_temp)
                print("\nGame Over.\n")                
                print(turn + " won.")
                break
            elif board_temp['B1'] == board_temp['B2'] == board_temp['B3'] != ' ': # down the middle
                show_board(board_temp)
                print("\nGame Over.\n")                
                print(turn + " won.")
                break
            elif board_temp['C1'] == board_temp['C2'] == board_temp['C3'] != ' ': # down the right side
                show_board(board_temp)
                print("\nGame Over.\n")                
                print(turn + " won.")
                break 
            elif board_temp['A1'] == board_temp['B2'] == board_temp['C3'] != ' ': # diagonal from NW to SE
                show_board(board_temp)
                print("\nGame Over.\n")                
                print(turn + " won.")
                break
            elif board_temp['A3'] == board_temp['B2'] == board_temp['C1'] != ' ': # diagonal from SW to NE
                show_board(board_temp)
                print("\nGame Over.\n")                
                print(turn + " won.")
                break 

        # If neither X nor O wins and the board is full, declare the result as a 'tie'.
        if count == 9:
            print("\nGame Over.\n")                
            print("It's a Tie!")

        # Now we have to change the player after every move.
        if turn =='X':
            turn = 'O'
        else:
            turn = 'X'
```

```python
play_game()
```
